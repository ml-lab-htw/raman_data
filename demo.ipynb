```python
from raman_data import raman_data
from matplotlib import pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
# List available datasets and plot the first two spectra from each dataset
datasets = raman_data()

datasets
```

```python

# Iterate through each dataset and plot the first two spectra
for dataset_name in datasets:
    print(f"Dataset: {dataset_name}")

    # Load the dataset
    dataset = raman_data(dataset_name)

    # Extract spectra and target labels
    data = dataset.data
    x_axis = dataset.spectra
    target = dataset.target
    metadata = dataset.metadata

    # Plot the spectra
    plt.plot(x_axis, data)
    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=20))
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.xlabel('Raman Shift')
    plt.ylabel('Intensity')
    plt.title(metadata["full_name"])
    #plt.legend()
    plt.show()
```
