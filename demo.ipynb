# Demo Raman Datasets

This notebook demonstrates the usage workflow of `raman_data` package and its available datasets.

```python
%load_ext autoreload
%autoreload 2
```

```python
# Step 1: Import the package and check available datasets
from raman_data import raman_data

datasets = raman_data()

print("Currently available datasets:")
datasets
```

```python
# Optionally you can specify the task type of a dataset
from raman_data import TASK_TYPE

classification_datasets = raman_data(task_type=TASK_TYPE.Classification)

print("Currently available classification datasets:")
classification_datasets
```

```python
# Step 2: Choose a dataset and analyse it
from matplotlib import pyplot as plt

chosen_datasets = ["codina/diabetes/AGEs", "Adenine"]

# Iterate through each dataset and plot the first two spectra
for dataset_name in chosen_datasets:
    print(f"Dataset: {dataset_name}")

    # Load the dataset
    dataset = raman_data(dataset_name)

    # Extract spectra and target labels
    spectra      = dataset.spectra       # intensity data
    raman_shifts = dataset.raman_shifts  # wavenumber values (x-axis)
    target       = dataset.target
    metadata     = dataset.metadata

    # Check additional information
    print(f"Dataset metadata:\n"
          f" Full name: {metadata['full_name'][:50]}...\n"
          f" Source: {metadata['source']}\n"
          f" Paper: {metadata['paper']}\n"
          f" Description: {metadata['description'][:50]}...")

    # Plot the first two spectra
    for i in range(2):
        plt.plot(raman_shifts, spectra[:, i])
        plt.xlabel('Raman Shift')
        plt.ylabel('Intensity')
        plt.title(f"{dataset_name} - Spectra {i + 1}")
        plt.show()
```

```python
# Another example to look at the whole dataset
chosen_dataset = "chlange/RamanSpectraEcoliFermentation"

# Load the dataset
dataset = raman_data(chosen_dataset)

# Extract spectra and target labels
spectra      = dataset.spectra       # intensity data
raman_shifts = dataset.raman_shifts  # wavenumber values (x-axis)

# Plot the dataset
plt.plot(raman_shifts, spectra[:, :5])
plt.xlabel('Raman Shift')
plt.ylabel('Intensity')
plt.title(f"{chosen_dataset} - whole spectra")
plt.xticks([_ for _ in range(0, 2000, 300)])
plt.show()
```
